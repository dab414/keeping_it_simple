## Disconnect Green Analysis

The task selection analyses for the disconnect_green data, *excluding* looking at block effects.

```r
library(dplyr)
library(data.table)
library(lme4)
library(ggplot2)
```

```r
current_data <- fread('../data/disconnect_green.csv')
head(current_data)
str(current_data)
```

```r
current_data$color <- as.factor(current_data$color)
```

Looking at the the observed subject and cell means of point difference predicting switching rates.

```r
options(repr.plot.width = 5, repr.plot.height = 4)
subject_means <- current_data[, .(transcode = mean(transcode)), by = .(subject, difference)]
subject_means$subject <- as.factor(subject_means$subject)
subject_means %>%
  group_by(difference) %>%
  summarize(switch = mean(transcode), se = sd(transcode) / sqrt(n())) %>%
  ggplot(aes(x = factor(difference), y = switch, group = 1)) + geom_line(size = 2) + 
geom_ribbon(aes(ymin = switch - se, ymax = switch + se), alpha = .2) + 
geom_line(data = subject_means, aes(x = factor(difference), y = transcode, group = subject, color = subject)) +
    ylim(0,1) + theme(legend.position = 'none') + xlab('Difference (Positive values indicate more reward for switching)') + 
ylab('Proportion of Switching')
```

Color ranges: 1-3 — Black, 4-6 — Green, 7-9 — Light Green

This is qualitatively a large contrast from the pattern we saw in Experiment 1. But the goal here should obviously be to describe and interpret these data in their own right, and worry about the cross-experiment logic in the following section.

At positive differences, we’re seeing what appears to be a pretty positive slope. The slope additionally appears to be modulated by the colors, where we see steep slopes in the black and light green ranges, and a less dramatic slope in the green range. The variability in the subject data looks more uniform — we’re no longer seeing the dramatically bimodal pattern that we saw in Experiment 1. Most subjects (with a couple exceptions) look like their responses mirror the cell means in the sense that their switch rates increase linearly as a function of the difference.

Let’s look at only the positive range on the difference variable, and run a linear regression with color nested within the difference.

```r
current_data <- current_data %>%
  mutate(color = as.factor(color),
         difference_c = difference - 5) %>%
  data.table()
## make sure experiment length doesn't alter effects
#current_data <- current_data[block < 11]
nest_model <- glmer(transcode ~ difference_c/color + (1 | subject), 
                    data = current_data[difference > 0], family = binomial(link = 'logit'))
```

```r
test_data <- data.table('difference_c' = -4:4)
  test_data <- cbind(test_data, green = c(rep(0,3),rep(1,3), rep(0,3)), light_green = c(rep(0, 6), rep(1, 3)))
  
  intercept <- fixef(nest_model)[1]
  b1 <- fixef(nest_model)[2]
  b2 <- fixef(nest_model)[3]
  b3 <- fixef(nest_model)[4]
  
  test_data <- test_data %>%
    mutate(prob = plogis(intercept + difference_c*b1 + green*b2 + light_green*b3))
  
  subject_means <- current_data %>%
    filter(difference > 0) %>%
    mutate(subject = as.factor(subject)) %>%
    group_by(subject, difference_c) %>%
    summarize(transcode = mean(transcode)) 
  
  subject_means$subject <- as.factor(subject_means$subject)
  
  cell_means <- subject_means %>%
    group_by(difference_c) %>%
    summarize(transcode = mean(transcode)) 
```

```r
summary(nest_model)
exp(fixef(nest_model))
```

```r
cell_means %>%
    ggplot(aes(x = factor(difference_c + 5), y = transcode, group = 1)) + 
          geom_line(size = 2) + 
          geom_line(data = subject_means, aes(x = factor(difference_c + 5), y = transcode, group = subject, 
                                              color = factor(subject))) + 
          geom_line(data = test_data, aes(x = factor(difference_c + 5), y = prob), size = 2, color = 'blue') + 
          theme(legend.position = 'none') + xlab('Difference (Amount of reward for switching)') + 
          ylab('Probability of Switching')
```

The blue line is the estimates from the model, the black line is the observed means.
